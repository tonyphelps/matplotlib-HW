```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
cityride_data = pd.merge(city_data, ride_data, how="outer", on=["city"])

# Display the data table for preview
cityride_data.head()
```

```python
cr_cities = cityride_data.groupby("city")

avg_fare_city = cityride_data.groupby("city")["fare"].mean()
rides_per_city = cityride_data.groupby("city")["ride_id"].count()
drivers_percity = cityride_data.groupby("city")["driver_count"].first()
city_type = cityride_data.groupby("city")["type"].first()

data_df = pd.DataFrame({
    "Avg_Fare" : avg_fare_city,
    "Rides" : rides_per_city,
    "Total_Drivers" : drivers_percity,
    "City_Type" : city_type})
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
urban_avg = data_df.loc[data_df.City_Type == "Urban", "Avg_Fare"]
suburban_avg = data_df.loc[data_df.City_Type == "Suburban", "Avg_Fare"]
rural_avg = data_df.loc[data_df.City_Type == "Rural", "Avg_Fare"]

urban_rides = data_df.loc[data_df.City_Type == "Urban", "Rides"]
suburban_rides = data_df.loc[data_df.City_Type == "Suburban", "Rides"]
rural_rides = data_df.loc[data_df.City_Type == "Rural", "Rides"]

urban_drivers = data_df.loc[data_df.City_Type == "Urban", "Total_Drivers"]*10
suburban_drivers = data_df.loc[data_df.City_Type == "Suburban", "Total_Drivers"]*10
rural_drivers = data_df.loc[data_df.City_Type == "Rural", "Total_Drivers"]*10

# Build the scatter plots for each city types
plt.scatter(urban_rides, urban_avg, s=urban_drivers, color = "coral", alpha=.6,edgecolors="black", linewidth=1, label="Urban")
plt.scatter(suburban_rides, suburban_avg, s=suburban_drivers, color = "lightskyblue", alpha=.5, edgecolors="black", linewidth=1,label="Suburban")
plt.scatter(rural_rides, rural_avg, s=rural_drivers,alpha=.5, color="gold", edgecolors="black", linewidth=1,label="Rural")

# Incorporate the other graph properties
plt.grid()
plt.title("Total Fares By City Type (2016)")
plt.xlabel("Total # Of Rides Per City")
plt.ylabel("Average Fare ($)")

# Create a legend
leg = plt.legend(title="City Types")

# Incorporate a text label regarding circle size
t = "Note:"\
    " Circle size correlates with driver count per city."
plt.text(43, 35, t, wrap=True)
# Save Figure
plt.savefig("../Images/FaresByCtypeScatter.png")
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
fares_pertype = cityride_data.groupby("type")["fare"].sum()
fare_total = fares_pertype.sum()
ctype_fpercent = ( fares_pertype / fare_total )
ctype_fpercent

# Build Pie Chart
colors = ["gold","lightskyblue","lightcoral"]
labels = ["Rural", "Suburban", "Urban"]
explode = (0.02, 0.03, 0.05)
plt.pie(ctype_fpercent,labels=labels, colors=colors, explode=explode, autopct="%1.1f%%", shadow=True, startangle=155)
plt.title("% of Total Fares by City Type")

# Save Figure
plt.savefig("../Images/FaresByCtypePie.png")
```

```python
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
ridez = cityride_data.groupby("type")["ride_id"].count()
rtotal = ridez.sum()
ctype_rpercent = (ridez / rtotal)
ctype_rpercent

# Build Pie Chart
colors = ["gold","lightskyblue","lightcoral"]
labels = ["Rural", "Suburban", "Urban"]
explode = (0.02, 0.03, 0.05)
plt.pie(ctype_rpercent,labels=labels, colors=colors, explode=explode, autopct="%1.1f%%", shadow=True, startangle=155)
plt.title("% of Total Rides by City Type")
# Save Figure
plt.savefig("../Images/TridesByCtypePie.png")
```

```python
# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
dcount = data_df.groupby("City_Type")["Total_Drivers"].sum()
dcount_total = dcount.sum()
ctype_dpercent = (dcount / dcount_total)
ctype_dpercent
# Build Pie Charts
colors = ["gold","lightskyblue","lightcoral"]
labels = ["Rural", "Suburban", "Urban"]
explode = (0.02, 0.03, 0.05)
plt.pie(ctype_dpercent,labels=labels, colors=colors, explode=explode, autopct="%1.1f%%", shadow=True, startangle=160)
plt.title("% of Total Drivers by City Type")
# Save Figure
plt.savefig("../Images/TdriversByCtypePie.png")
```

```python
# Show Figure
plt.show()
```
